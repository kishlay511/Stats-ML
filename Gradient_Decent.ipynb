# Gradient Descent Function: 

### Finding the best fit line on the given parameter
### Calculating Slope, Intercept, Cost Function
### Cost Function of Gradient Descent

```python

import numpy as np
def gradient_descent (x,y):
    m = 0
    c = 0

    learning_rate = 0.01
    iterations = 5000 
    n = len(x)

    for i in range(iterations):

        y_predicted = m * x + c
    
        cost = (1/n) * sum([val**2 for val in ( y - y_predicted)])
    
        dm = (-2/n) * sum(x * (y - y_predicted))
        dc = (-2/n) * sum (y - y_predicted)
    
        m = m - learning_rate * dm

        c = c - learning_rate * dc

        print("slope = {}, intercept = {}, iterations = {}, cost {}".format(m,c,i+1, cost))
```

```python
x = np.array([1,2,3,4,5]) 
y = np.array([5,6,7,6,5])
```

```python
gradient_descent (x,y)
```
